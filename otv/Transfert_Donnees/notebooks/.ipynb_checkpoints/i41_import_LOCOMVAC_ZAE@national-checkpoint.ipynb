**Fichier des locaux commerciaux vacants**

https://www.collectivites-locales.gouv.fr/finances-locales/fichier-des-locaux-commerciaux-vacants

En application de [l'article 1530](https://www.legifrance.gouv.fr/codes/article_lc/LEGIARTI000026948469/2022-04-06/) du Code général des impôts (CGI), les communes ou les EPCI ayant une compétence en matière d'aménagement des zones d'activité commerciale peuvent instituer une taxe annuelle sur les friches commerciales.

Cette taxe s'applique aux locaux commerciaux qui ne sont plus affectés à une activité entrant dans le champ d'application de la cotisation foncière des entreprises (CFE) défini à [l'article 1447 du CGI](https://www.legifrance.gouv.fr/codes/article_lc/LEGIARTI000023380872/2022-04-06/). Au 1er janvier de l'année d'imposition, l’absence d’exploitation doit avoir perduré pendant au moins deux ans et doit être dépendante de la volonté du contribuable.


Le fichier « LOCOMVACxx » dresse la liste des locaux commerciaux et professionnels dépourvus d'imposition à la CFE l'année précédente sur le périmètre de la collectivité concernée.

Ce fichier est mis à disposition des collectivités en application de [l’article L. 135 B](https://www.legifrance.gouv.fr/codes/article_lc/LEGIARTI000033219216/2022-04-06/) du livre des procédures fiscales.

Le fichier est ainsi adressé aux collectivités (communes, y compris lorsqu'elles sont membres d'un EPCI à fiscalité unique, et EPCI à fiscalité additionnelle ou unique). Il constate pour les locaux désignés une absence d'imposition à la CFE pour un millésime donné. Il ne permet pas de préjuger de manière certaine, dans toutes les situations, de l’absence d'occupation du local.

En effet, un local vacant au sens de la CFE peut toutefois être occupé par un usager professionnel dont l'activité se situe hors du champ d'application de la CFE ou est exonérée de cette cotisation à titre permanent. Dès lors, les données des fichiers requièrent un examen et un retraitement par les collectivités avant toute transmission aux services des finances publiques des informations nécessaires à l'établissement de la taxe sur les friches commerciales.

Depuis 2018, la mise à disposition des listes de locaux a lieu habituellement en avril.

**Mise à disposition des collectivités locales sur le PIGP**

Un mail est envoyé à la collectivité locale afin de l'informer de la disponibilité du fichier. La collectivité doit se connecter au Portail internet de la gestion publique (PIGP) pour télécharger le fichier. Elle dispose à cette fin d’un délai de 30 jours. Une relance automatique est faite au bout de 11 jours si aucun téléchargement n'a été effectué et, si nécessaire, une deuxième relance est faite au 22e jour. Au bout de 30 jours, le système purge les informations et le fichier devient inaccessible.

Les listes de locaux vacants au sens de la CFE au titre de 2021 sont mises à la disposition des collectivités locales sur le PIGP au cours du mois d’avril 2022.

**Caractéristiques du fichier**

Le fichier téléchargé est compressé au format zip. La décompression peut s'opérer avec les outils habituels : 7-zip, winrar, winzip. Une fois décompressé, le fichier porte le même nom, à l'exception du suffixe d'extension.

Le fichier est de type texte, au format CSV, c'est-à-dire qu'un séparateur (ici le « ; ») distingue chaque champ (voir le descriptif du fichier plus loin).

Si le fichier téléchargé ne comprend pas l'extension « .csv », il convient de la rétablir en modifiant le nom du fichier par l'ajout de « .csv » (sans les guillemets). Après cette opération, le fichier a la forme suivante : LOCOMVAC17_320_21320094200018_C094_18123.csv (exemple).

Il peut ensuite être ouvert dans un tableur (Microsoft Excel, LibreOffice Calc, etc.). Il est nécessaire, à l'ouverture de l'assistant, de préciser le séparateur unique.

**Information sur la Cotisation Foncière des Entreprises**

https://entreprendre.service-public.fr/vosdroits/F23547

La cotisation foncière des entreprises doit être réglée par les sociétés et par les entrepreneurs individuels (dont micro-entrepreneurs), y compris ceux exerçant leur activité à domicile ou chez leurs clients.

Pour être redevable de la CFE, l'activité du contribuable doit remplir les conditions suivantes :

    L'activité est exercée en France

    L'activité présente un caractère habituel : elle est exercée de manière répétée.
    L'activité est exercée à titre professionnel : cela exclut les activités sans but lucratif et les personnes qui se limitent à la gestion de leur patrimoine privé.
    L'activité est non salariée : les salariés ne sont pas concernés par la CFE

Les sociétés et entrepreneurs individuels doivent payer la CFE quels que soient leur statut juridique, la nature de leur activité, leur régime d'imposition et leur nationalité.


Les activités de location ou de sous-location d'immeubles nus sont concernées par la CFE lorsqu'elles génèrent un chiffre d'affaires ou des recettes brutes hors taxes d'au moins 100 000 €.

En revanche, la CFE ne concerne pas les activités de location et de sous-location d'immeubles nus à usage d'habitation. La CFE ne concerne pas non plus la location meublée de certains logements qui font partie de la résidence principale du propriétaire.


**Les contribuables peuvent être exonérés du paiement de la CFE de plein droit (automatiquement) ou de manière facultative (sur demande, avec approbation de la collectivité).**

À savoir  
**une entreprise est exonérée de CFE l'année de sa création**. Ensuite, sa base d'imposition est réduite de moitié l'année suivante.

**Exonérations facultatives**

    - Zones franches urbaines de seconde génération
    - Zones franches urbaines-territoires entrepreneurs de troisième génération
    - Zones de restructuration de la défense (ZRD : Territoire affecté par le redéploiement des implantations militaires, notamment par une baisse d'activité des entreprises. Les entreprises qui s'y installent peuvent bénéficier d'aides fiscales et sociales.)
    - Bassins d'emploi à redynamiser (BER)

Pour bénéficier d'une exonération facultative de CFE, l'entreprise doit effectuer une déclaration n°1447-M-SD (dite déclaration modificative)à adresser au service des impôts des entreprises (SIE) dont elle dépend.

```python

def lancementTache(reqs,tpslong=50,suivi=2,PgConn=PgConn):
    """
    prend en paramètre une requête sql ou un tableau requête sql
    tpslong: durée à partir de laquelle la requête est considérée comme excessivement longue et mise dans les logs
    suivi des logs:
        si l'une des lignes commencent par -->
        il affiche la ligne dans le log
    suivi: 
        0 : pas d'affichage dans INFO des -->
        1 : affichage uniquement du premier
        2 : affichae à chaque étape
    
    """
    def lancementTache1(req,suiviB=False):
        from time import time
        t0=time()
        taches=[x for x in req.split('\n') if '-->' in x]
        if taches!=[]:
            if suiviB:
                logger.info(taches)
        logger.debug(req)
        try:
            PgConn.execute(req)
            logger.debug(f"FINI:{req} ")
        except Exception as e:
            logger.error(f"Erreur SQL {e}")        
        if taches!=[]:
            t1=time()
            if (t1-t0)>tpslong:
                logger.warning(f" {taches} terminée durée: {round(t1-t0)} s > {tpslong} ")
            else:
                logger.debug(f" {taches} terminée durée: {round(t1-t0)} s")
    
    if isinstance(reqs, (list, tuple)):
        # c'est un tableau de requête
        for (i,req1) in enumerate(reqs):
            if (suivi==2) | (suivi==1 & i==0):
                lancementTache1(req1,suiviB=True)
            else:
                lancementTache1(req1,suiviB=False)
    else:
        if (suivi==2) | (suivi==1):
            lancementTache1(reqs,suiviB=True)
        else:
            lancementTache1(reqs,suiviB=False)
```

```python
def copy_to_sql(fichier,table,size=8192,encoding='iso-8859-1',sep=';',quote='"'):
    conn = psycopg2.connect(f"host={f.server} port={f.port} dbname={f.dbname} user={f.user} password={f.password}")
    conn.set_isolation_level(0)
    cur = conn.cursor()
    cur.execute(f"SET CLIENT_ENCODING TO '{encoding}'")
    SQL_STATEMENT = f"""COPY {table} FROM STDIN WITH CSV HEADER DELIMITER AS '{sep}' QUOTE AS '{quote}' """
    with open(fichier, 'r') as ff:
        cur.copy_expert(sql=SQL_STATEMENT, file=ff, size=size)
    cur.close()
```

```python
if instance=='V0':
    """
     pour chaque uf ayant un local vacant 
     calculer un taux de vacance en nb de locaux, en surface de locaux, en surface de pev
     pour les locaux vacants on écarte certains types de vacance '0'
    """

    req4=[f"""--> calcul des aggrégats de surface sur les UFs
    DROP TABLE IF EXISTS {schemadest}.ufagg_loccomvac_tmp;
    CREATE TABLE IF NOT EXISTS {schemadest}.ufagg_loccomvac_tmp as
    with uf as (
    select distinct id_uf from plug_zae.rel_uf_loccom_vac_tmp
    )
    SELECT 
    uf.id_uf,
    sum(nloghvac) nloghvac, 
    sum(nactvacant) nactvacant, 
    sum(nloghvac2a) nloghvac2a, 
    sum(nactvac2a) nactvac2a, 
    sum(nloghvac5a) nloghvac5a, 
    sum(nactvac5a) nactvac5a, 
    sum(npevp) npevp, 
    sum(sprincp) sprincp, 
    sum(ssecp) ssecp, 
    sum(ssecncp) ssecncp, 
    sum(sparkp) sparkp, 
    sum(sparkncp) sparkncp, 
    sum(slocal) slocal 
    FROM ff{millesimeF}.fftp_{millesimeF}_pnb10_parcelle f,
    ff_fr_{millesimeF}.idpar p, uf
    where p.id_uf = uf.id_uf and p.idpar=f.idpar
    group by uf.id_uf    
    """,f"""
    alter table {schemadest}.ufagg_loccomvac_tmp add primary key (id_uf);
    """,f"""
    -- {schemadest}.uf_loccomvac_tmp definition
    -- Drop table
    DROP TABLE IF EXISTS {schemadest}.uf_loccomvac_tmp;
    CREATE TABLE IF NOT EXISTS {schemadest}.uf_loccomvac_tmp (
        id_ufvac varchar,
        annee varchar(4),
        ff_sact_vac int8 NULL,
        ff_slocal_vac int8 NULL,
        nb_bat_vac int8 NULL,
        pourc_sact_vac int8 NULL,
        pourc_slocal_vac int8 NULL,
        nloghvac int8 NULL,
        nactvacant int8 NULL,
        nloghvac2a int8 NULL,
        nactvac2a int8 NULL,
        nloghvac5a int8 NULL,
        nactvac5a int8 NULL,
        nmediocre int8 NULL,
        debutvacancemin varchar NULL,
        debutvacancemax varchar NULL, 
        adresse text NULL,
        typo_vacdomin varchar NULL,
        insee varchar(5)
    );

    COMMENT ON TABLE {schemadest}.uf_loccomvac_tmp IS 'unités foncières et locaux d''activité sans perception de CFE';

    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.id_ufvac IS 'identifiant d''uf';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.annee IS 'millésime LOVAC';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.ff_sact_vac IS 'surface des parties principales professionnelles et des parties secondaires couvertes professionnelles (en m2) sans paiement de la CFE à l''échelle de l''uf';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.ff_slocal_vac IS 'surface totale des parties d''évaluation vacantes à l''échelle de l''uf';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.pourc_sact_vac IS 'part vacante de la surface des parties principales professionnelles et des parties secondaires couvertes professionnelles à l''échelle de l''uf (en %%)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.pourc_slocal_vac IS 'part vacante de la surface des parties d''évaluation à l''échelle de l''uf (en %%)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nloghvac IS 'nombre de logements vacants sur l''uf (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nactvacant IS 'nombre de locaux d''activité vacants sur l''uf (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nactvac2a IS 'nombre de locaux d''activité vacants depuis plus de 2 ans sur l''uf (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nactvac5a IS 'nombre de locaux d''activité vacants depuis plus de 5 ans sur l''uf (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nloghvac2a IS 'nombre de logements vacants depuis plus de 2 ans à l''échelle de l''uf (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nloghvac5a IS 'nombre de logements vacants depuis plus de 5 ans à l''échelle de l''uf (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nmediocre IS 'nombre de logements médiocres (source FF)';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.nb_bat_vac IS 'nombre de bâtiments possédant un logement vacant à l''échelle de l''uf';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.debutvacancemin IS 'au moins un logement vacant depuis';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.debutvacancemax IS 'tous les locaux d''activité sans perception de CFE avant';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.adresse IS 'adresse des locaux d''activité sans perception de CFE';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.insee IS 'COG communal';
    COMMENT ON COLUMN {schemadest}.uf_loccomvac_tmp.typo_vacdomin IS 'Typologie de vacance dominante de locaux';
    
    -- Permissions
    ALTER TABLE {schemadest}.uf_loccomvac_tmp OWNER TO urban;
    GRANT ALL ON TABLE {schemadest}.uf_loccomvac_tmp TO urban;
    """,f"""
    INSERT INTO {schemadest}.uf_loccomvac_tmp (id_ufvac, annee, ff_sact_vac, ff_slocal_vac, nb_bat_vac,
     pourc_sact_vac, pourc_slocal_vac,
    nloghvac, nactvacant, nloghvac2a, nactvac2a, nloghvac5a, nactvac5a, nmediocre,
    --debutvacancemin, debutvacancemax,
    adresse, insee, typo_vacdomin
    )
    select id_uf, ufv.annee, ff_sact_vac, ff_slocal_vac, nb_bat_vac, 
    case when (sprincp+ssecp)!=0 then ff_sact_vac::double precision/(sprincp+ssecp)::double precision*100.0 else 0 end as pourc_sact_vac,
    case when slocal!=0 then ff_slocal_vac::double precision/slocal::double precision*100.0 else 0 end as pourc_slocal_vac,
    nloghvac, nactvacant, nloghvac2a, nactvac2a, nloghvac5a, nactvac5a, NULL as nmediocre,
    --debutvacancemin, debutvacancemax, 
    ufv.adresse, substring(id_uf,1,5) as insee, ufv.typo_vacdomin
    from
    (select id_uf, l.annee, sum(ff_sprincp+ff_ssecp) as ff_sact_vac,
     sum(coalesce(ff_slocal,0)) as ff_slocal_vac,
     count(distinct ff_idbat)  as nb_bat_vac,
     --MIN(debutvacance) as debutvacancemin, MAX(debutvacance) as debutvacancemax,
     array_to_string(array_agg(distinct coalesce(num_voirie,'')||' '||coalesce(cod_indic_repet,'')||' '||coalesce(adresse,'')||' '||coalesce(code_postal,'')||' '||coalesce(libcom,'')),';')::text as adresse,
      mode() WITHIN GROUP (ORDER BY typo_vac) as typo_vacdomin
     from 
        {schemadest}.rel_uf_loccom_vac_tmp ufv, 
        {schemadest}.{tabledest} l 
        where ufv.gid_loccom_vac=l.gid 
        and l.annee='{anneeLOCOMVAC}'
        and (typo_vac NOT IN ('0') or typo_vac isnull)
        group by id_uf, l.annee) as ufv
    left join 
    {schemadest}.ufagg_loccomvac_tmp using(id_uf)
    -- est ce que l'on fait une sélection sur la dominante de locaux
    --where greatest(nlocmaison, nlocappt, nloccom)=nlocappt 
    """,
    f"""
    alter table {schemadest}.uf_loccomvac_tmp add primary key (id_uf);
    """]

    lancementTache(req4,suivi=2,PgConn=PgConnProd)
```
